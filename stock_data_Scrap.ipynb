# Fatching  Multiple Tickers(company) Data

```python
import yfinance as yf
import pandas as pd

# Define a list of ticker symbols
ticker_symbols = ['ZOMATO.NS', 'INFY.NS', 'TCS.NS']  
start_date = '2019-01-01'
end_date = '2024-08-20'

for ticker_symbol in ticker_symbols:
    # Fetch historical stock data
    data = yf.download(ticker_symbol, start=start_date, end=end_date)
    
    # Reset the index to have Date as a column
    data.reset_index(inplace=True)
    
    filename = f'{ticker_symbol}_historical_stock_data.csv'
    data.to_csv(filename, index=False)
    
    print(f"Data for {ticker_symbol} fetched and saved successfully as {filename}.")
```

```python
infosys = pd.read_csv("INFY.NS_historical_stock_data.csv")
tcs = pd.read_csv("TCS.NS_historical_stock_data.csv")
zomato = pd.read_csv("ZOMATO.NS_historical_stock_data.csv")
```
