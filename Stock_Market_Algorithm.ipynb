# Indian Stock Market Datas useing Yahoo Finanace

```python
import yfinance as yf
import pandas as pd

ticker_symbol = 'INFY.BO'

# Fetch data for the most recent day
stock = yf.Ticker(ticker_symbol)
data = stock.history(period='1d')

# Extract details
latest_data = data.iloc[-1]
open_price = latest_data['Open']
high_price = latest_data['High']
low_price = latest_data['Low']
close_price = latest_data['Close']
volume = latest_data['Volume']

# Calculate previous close and change
previous_close = data['Close'].iloc[-2] if len(data) > 1 else None
change = close_price - previous_close if previous_close is not None else None
percent_change = (change / previous_close * 100) if previous_close is not None else None

# 52-week data
history_52_week = stock.history(period='1y')
low_52_week = history_52_week['Low'].min()
high_52_week = history_52_week['High'].max()
low_change = close_price - low_52_week
high_change = close_price - high_52_week
low_change_percent = (low_change / low_52_week * 100) if low_52_week != 0 else None
high_change_percent = (high_change / high_52_week * 100) if high_52_week != 0 else None

# Output the results
print(f"Real-Time Stock Data for {ticker_symbol}")
print(f"Currency: INR")
print(f"Date & Time: {latest_data.name.date()}")
print(f"Open Price (INR): {open_price:.2f}")
print(f"High Price (INR): {high_price:.2f}")
print(f"Low Price (INR): {low_price:.2f}")
print(f"Close Price (INR): {close_price:.2f}")
print(f"Volume: {volume}")
print(f"Previous Close: {previous_close:.2f}" if previous_close else "Previous Close: N/A")
print(f"Change: {change:.2f}" if change else "Change: N/A")
print(f"Percent Change: {percent_change:.2f}" if percent_change else "Percent Change: N/A")
print(f"Average Volume: {data['Volume'].mean():.0f}")
print(f"Is Market Open: {'Yes' if data.index[-1].date() == pd.Timestamp.now().date() else 'No'}")
print(f"52-Week Low: {low_52_week:.2f}")
print(f"52-Week High: {high_52_week:.2f}")
print(f"52-Week Low Change: {low_change:.2f}")
print(f"52-Week High Change: {high_change:.2f}")
print(f"52-Week Low Change Percent: {low_change_percent:.2f}")
print(f"52-Week High Change Percent: {high_change_percent:.2f}")
print(f"52-Week Range: {low_52_week:.2f} - {high_52_week:.2f}")
```
